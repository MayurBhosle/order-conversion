# Cart Abondanment Rate per Customer

```python
import pandas as pd

# dataset - https://www.kaggle.com/mkechinov/ecommerce-events-history-in-cosmetics-shop?select=2019-Nov.csv

df=pd.read_csv('C:\order-conversion\\2019-Nov.csv')
df=df[df['event_type'] != 'view']  #removing the view cart event
df
```

```python
#extract the users who added some product in the cart
cart_users=df[(df['event_type'] == 'cart')].user_id.unique()
cart_users
```

```python
#removing the rows of the users who didn't anything to the cart
df= df[(df['user_id'].isin(cart_users))]
df
```

```python
def customer_cart_abondan(user_id):
    user_df= df[df['user_id']== user_id]
    
    added_to_cart=user_df[user_df['event_type']=='cart']
    added_to_cart=added_to_cart.product_id
    removed_from_cart=user_df[user_df['event_type']=='remove_from_cart'].product_id
    
    left_in_cart=added_to_cart[~added_to_cart.isin(removed_from_cart)]
    left_in_cart_count= left_in_cart.shape[0]
    purchased=user_df[user_df['event_type']=='purchase'].product_id
    
    abondaned=left_in_cart[~left_in_cart.isin(purchased)]
    abondand_count=abondaned.shape[0]
    return [user_id,abondand_count,left_in_cart_count]
```

```python
import time
begin=time.time()
arr=[]
# a=0
for i in cart_users:
    arr.append(customer_cart_abondan(i))
#     a=a+1
#     print(a)
end=time.time()
print(end-begin)
```

```python
arr
```

```python
final_df= pd.DataFrame(arr, columns=['user_id','aboundand_items', 'items_left_after_remove'])
final_df['rate']=(final_df['aboundand_items']*100)/final_df['items_left_after_remove']
```

```python
final_df
```

```python
final_df.sort_values('rate')
```

# Finding the No of aboundand sessions


```python
import pandas as pd

# dataset - https://www.kaggle.com/mkechinov/ecommerce-events-history-in-cosmetics-shop?select=2019-Nov.csv

df=pd.read_csv('C:\order-conversion\\2019-Nov.csv')
```

```python
df=df[df['event_type']!='view']
df=df[['user_id', 'product_id', 'event_type','user_session']]
df
```

```python
cart_sessions=df[df['event_type']=='cart'].user_session.unique()
cart_sessions.shape
```

```python
df=df[(df['user_session'].isin(cart_sessions))]
df
```

```python
def customer_cart_abondan(user_id): # 
    user_df=df[df['user_id']==user_id]
    sessions=user_df.user_session.unique()
    arr=[]
    for sess in sessions:
        session_df=user_df[user_df['user_session']==sess]
        added_to_cart=session_df[session_df['event_type']=='cart'].product_id
        removed_from_cart=session_df[session_df['event_type']=='remove_from_cart'].product_id
        left_in_cart=added_to_cart[~added_to_cart.isin(removed_from_cart)]
        left_in_cart_count= left_in_cart.shape[0]
        purchased=session_df[session_df['event_type']=='purchase'].product_id
        abondaned=left_in_cart[~left_in_cart.isin(purchased)]
        abondand_count=abondaned.shape[0]
        if(abondand_count > 0):
            arr.append([user_id,sess, "abandoned" ])
        else:
            arr.append([user_id,sess, "Not abandoned" ])

    return arr
```

```python
users=df.user_id.unique()
users.shape
```

```python
import time as time
arr=[]
# a=0
begin=time.time()
for i in users:
    arr.extend(customer_cart_abondan(i))
#     a=a+1
#     print(a)
#     if(a==100):
#         break
end=time.time()
print(end-begin)
```

```python
arr
```

```python
dff=pd.DataFrame(arr,columns=['user_id','session','abandoned'])
dff
```

```python
dff.groupby(['user_id','abandoned']).count().sort_values(['abandoned','session'],ascending=[False,False])
```
